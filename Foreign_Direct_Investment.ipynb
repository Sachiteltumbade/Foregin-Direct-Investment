```python
#Foreign Direct Investment
##importing important librabies for data manipulation and analysis and for data visualization

import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
#reading csv file from drive
FDI =  pd.read_csv('/content/drive/MyDrive/FDI data.csv')
```

```python
FDI.style.set_caption("Amount in US$ Millions"). format(precision=2)
```

```python
#returns the column labels as an index object
FDI.columns
```

```python
Year = ['2000-01', '2001-02', '2002-03', '2003-04','2004-05','2005-06','2006-07','2007-08','2008-09',
        '2009-10','2010-11','2011-12','2012-13','2013-14','2014-15','2015-16','2016-17'] #representing the financial years from 2000-01 to 2016-17.

Sectors = ['Sector']
```

```python
FDI.info() ##prints summary of the dataframe
```

```python
FDI.isnull().sum() ##returns the number of missing values in each column
```

```python
Rates = [45.68, 47.69, 48.39, 45.95, 44.93, 44.27, 45.24, 40.26, 45.99,
         47.44, 45.56, 47.92, 54.40, 60.50, 61.14, 65.46, 67.07]
```

```python
#takes dataframe,list of column, and a number as an input
def multiply_columns(df, col_list, num):
  for col in col_list:
    df[col] = df[col]* Rates[col_list.index(col)]/10
  return df
```

```python
FDI_InUSD= FDI.copy() #making copyof the DataFrame
FDI_02 = multiply_columns(FDI, Year, Rates)
```

```python
FDI_02.style.set_caption("FDI INFLOWS(Amount in  Crores)").format(precision=2) ##Accesses the styling functionality of the DataFrame
```

```python
##reshapig the dataframe
melt = pd.melt(FDI_InUSD, id_vars = Sectors, value_vars = Year, var_name = 'Year',
               value_name='FDI(US$ Million)',ignore_index=True)
melt
```

```python
melt01 = pd.melt(FDI_02, id_vars = Sectors, value_vars = Year, var_name='Year',
                 value_name = 'FDI( Crores)', ignore_index=True)
melt01=round(melt01,2)
melt01
```

```python
Merged = melt01.merge(melt, how='left')
Merged
```

```python
Sorted = Merged.sort_values(['Sector', 'Year'],ignore_index=True)
Sorted
```

```python
print("\nStats for Sectors\n",'-'*65, sep='')
print(pd.DataFrame(Sorted.groupby('Sector').describe().loc[:,:]).transpose())
```

```python
Sorted = Sorted[['Sector','FDI( Crores)','FDI(US$ Million)'
                 ,]].replace(["CONSTRUCTION DEVELOPMENT: Townships, housing, build-up infrastructure and construction-development projects",
                             "SERVICES SECTOR (Fin., Banking, Insurance, Non Fin/Business, Outsourcing, R&D, Courier, Tech. Testing and Analysis, Other)",
                             "TEA AND COFFEE (PROCESSING & WAREHOUSING COFFEE & RUBBER)"],
                            ["CONSTRUCTION DEVELOPMENT","SERVICES SECTOR","TEA AND COFFEE"])
```

```python
Sectorwise_fdi = Sorted.groupby('Sector').sum()
Sectorwise_fdi.sort_values(by='FDI(US$ Million)',ascending=False)
```

```python
Sectorwise_fdi.plot(kind='bar',y='FDI( Crores)',figsize = (25,7), color='red', legend=True,title='SECTOR_WISE FDI INFLOWS',ylabel='FDI( Crores)') #geenrating a bar plot that visually represents the FDI inflows in  Crores for different sectors, with customization options for size, color, legend, title, and axis labels
```

```python
Sectorwise_fdi.plot(kind='bar',y='FDI(US$ Million)',figsize = (25,7), color='yellow', legend=True,title='SECTOR_WISE FDI INFLOWS',ylabel='FDI(US$ Million)')  ###generating a bar plot that visually represents the FDI inflows in US$ Million for different sectors, with customization options for size, color, legend, title, and axis labels
```

```python
##extracts the top 10 sectors with the highest FDI inflows in  Crores from your Sectorwise_fdi
Top_10_Sectors = Sectorwise_fdi.nlargest(10,[ 'FDI( Crores)'])
```

```python
Total_fdi = round(melt01['FDI( Crores)'].sum(),2)
Sum = Top_10_Sectors['FDI( Crores)'].sum()
Top_10_Sectors['In %age'] = round(Top_10_Sectors['FDI( Crores)']/Sum*100,2)
Top_10_Sectors['%age to Total Inflows']= round(Top_10_Sectors['FDI( Crores)']/Total_fdi*100,2) # Moved closing parenthesis from after 100 to after Total_fdi
Top_10_Sectors
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(15,5))
plt.barh(Top_10_Sectors.index,Top_10_Sectors['FDI( Crores)'],color="orange") #generating a horizontal bar plot
plt.title('SECTORS ATTRACTING HIGHEST INFLOWS')
plt.xlabel('FDI( Crores)')
plt.ylabel('Top Sectors')
plt.show()
```

```python
##creating a pie chart to visualize the share of FDI inflows among the top 10 sectors.
plt.figure(figsize=(20,8))
plt.pie(Top_10_Sectors['FDI( Crores)'],labels=Top_10_Sectors.index,autopct='1%.1f%%',shadow=True,startangle=90)
plt.axis('equal')
plt.title('SHARE AMONG TOP 10SECTORS ATTRACTING HIGHEST FDIINFLOWS')
plt.show()
```

```python
Bottom_5_Sectors = Sectorwise_fdi.nsmallest(5,['FDI( Crores)'])
Sum = Bottom_5_Sectors['FDI( Crores)'].sum()
Bottom_5_Sectors['In %age'] = round(Bottom_5_Sectors['FDI( Crores)']/Sum*100,2)
Bottom_5_Sectors['%age to Total Inflows']= round(Bottom_5_Sectors['FDI( Crores)']/Total_fdi*100,3)
Bottom_5_Sectors
```

```python
#Visualization
#bar plot
plt.figure(figsize=(15,8))
plt.barh(Bottom_5_Sectors.index,Bottom_5_Sectors['FDI( Crores)'],color="green") #Creates the horizontal bar plot, Bottom_5_Sectors['FDI( Crores)'] uses the FDI values in  Crores as the x-axis values, and Bottom_5_Sectors.index uses the sector names
plt.title('SECTORS ATTRACTING LOWEST INFLOWS')
plt.xlabel('FDI( Crores)')
plt.ylabel('Bottom Sectors')
plt.show()
```

```python
#creating a pie chart to visualize the share of FDI inflows among the bottom 5 sectors
plt.figure(figsize=(20,8))
plt.pie(Bottom_5_Sectors['FDI( Crores)'], labels=Bottom_5_Sectors.index,autopct='1%.1f%%',shadow=True,startangle=90)
plt.axis('equal')
plt.title('SHARE AMONG BOTTOM 5 SECTORS ATTRACTING LOWEST FDIINFLOWS')
plt.show()
```
